```python
import torch
import torchaudio
```

```python
a, sample_rate = torchaudio.load(
    "/data/HEMC_klk_wavs/ID2365/ID2365_BL.wav",
)
print(a.shape)
print(f"{torch.max(a)}max")
print(f"{torch.min(a)}min")
print(f"{torch.mean(a)}mean")
print(f"{a.shape[1] / sample_rate}s")
print(f"{a.shape[1] / sample_rate / 3600}h")
print(f"{a.element_size() * a.nelement() / 1024 / 1024}mb")
print(f"{a.element_size() * a.nelement() / 1024 / 1024*4}mb total")
```
